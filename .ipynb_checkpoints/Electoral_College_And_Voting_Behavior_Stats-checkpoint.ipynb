```r
library(ggplot2)
library(plyr)
df = read.csv('data/stats_by_state.csv', header=TRUE)
# rename(df, c("X"="state"))
# df.to_csv('')
```

```r
# head(df)
# colnames(df)
# df['state']
# df['state']
newdata <- df[ which(df$state=='Puerto Rico'), ]
```

```r
fit = lm(ec_votes_per_voter_pct_ave ~ ave_age_eligible_white + 
                                      pct_white_eligible + 
                                      pct_rural + 
                                      pct_rep + 
                                      pct_dem_rep_gap,
                                      data = df)
summary(fit)
```

```r
fit = lm(ec_votes_per_voter_pct_ave ~ ave_age_eligible_white, data = df)
summary(fit)

ggplot(df[which(df$state != 'Puerto Rico'),],
       aes(x=ave_age_eligible_white,
           y=ec_votes_per_voter_pct_ave,
           color=winner))+
  geom_point()
```

```r
fit = lm(ec_votes_per_voter_pct_ave ~ ave_age_eligible, data = df)
summary(fit)

ggplot(df[which(df$state != 'Puerto Rico'),],
       aes(x=ave_age_eligible,
           y=ec_votes_per_voter_pct_ave,
           color=winner))+
  geom_point()
```

```r
fit = lm(ec_votes_per_voter_pct_ave ~ pct_dem_rep_gap_abs, 
         data = df[which(df$state != 'Puerto Rico'),])
summary(fit)

ggplot(df[which(df$state != 'Puerto Rico' & df$pct_dem_rep_gap_abs < .45),],
       aes(x=pct_dem_rep_gap_abs,
           y=ec_votes_per_voter_pct_ave,
           color=winner))+
  geom_point()
```

So the only things that are associated with Electoral College Over-Representation (besides small state populations) are:
1. Younger white voters (but not eligible voters in general)
1. A more partisan electorate (ie. an electorate where a larger majority of votes went to one candidate or the other)

The latter isn't surprising since smaller states are more likely to have a homogenous group of people, who are more likely to hold similar political views.
